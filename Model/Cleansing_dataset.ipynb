```python
import pandas as pd
import numpy as np
```

```python
df_recipes = pd.read_csv('/content/drive/MyDrive/Capstone/recipes.csv')
df_reviews = pd.read_csv('/content/drive/MyDrive/Capstone/reviews.csv')
```

```python
# Menampilkan total baris
total_rows = df_recipes.shape[0]
print(f"Total baris dalam dataset: {total_rows}")
```

```python
df_recipes.head(5)
```

```python
df_recipes.info()
```

```python
# Menghapus baris yang tidak memiliki link gambar
df_recipes = df_recipes[df_recipes['Images'].notna()]
# Menghapus baris yang memiliki ReviewCount di bawah 10
df_recipes = df_recipes[df_recipes['ReviewCount'] >= 10]
# Menghapus kolom-kolom yang diminta
columns_to_drop = ['AuthorName', 'CookTime', 'PrepTime', 'TotalTime', 'DatePublished', 'RecipeYield', 'Keywords', 'RecipeServings', 'Keywords', 'RecipeYield']
df_recipes = df_recipes.drop(columns=columns_to_drop, errors='ignore')
```

```python
df_recipes = df_recipes.rename(columns={
    'FatContent': 'Fat',
    'SaturatedFatContent': 'SaturatedFat',
    'CholesterolContent': 'Cholesterol',
    'SodiumContent': 'Sodium',
    'CarbohydrateContent': 'Carbohydrate',
    'FiberContent': 'Fiber',
    'SugarContent': 'Sugar',
    'ProteinContent': 'Protein'
})
```

```python
df_recipes['Description'] = df_recipes['Description'].replace('from Food.com', '', regex=True)
```

```python
df_recipes.head(5)
```

```python
# Tampilkan baris di mana 'RecipeIngredientQuantities' mengandung huruf NA
df_with_na = df_recipes[df_recipes['RecipeIngredientQuantities'].str.contains('NA')]

print(df_with_na.shape[0])
```

```python
# Hapus baris di mana 'RecipeIngredientQuantities' mengandung huruf NA
df_recipes = df_recipes[~df_recipes['RecipeIngredientQuantities'].str.contains('NA')]

df_recipes.head(5)
```

```python
# Menampilkan total baris
total_rows = df_recipes.shape[0]
print(f"Total baris dalam dataset: {total_rows}")
```

------------------------------------------

```python
# Menampilkan total baris
total_rows = df_reviews.shape[0]
print(f"Total baris dalam dataset: {total_rows}")
```

```python
df_reviews = pd.merge(df_reviews, df_recipes[['RecipeId', 'Name']], on='RecipeId', how='left')
```

```python
name_column = df_reviews.pop('Name')
df_reviews.insert(1, 'Name', name_column)
```

```python
# Menghapus baris yang tidak memiliki link gambar
df_reviews = df_reviews[df_reviews['Name'].notna()]
```

```python
df_reviews.head(5)
```

```python
# Menampilkan total baris
total_rows = df_reviews.shape[0]
print(f"Total baris dalam dataset: {total_rows}")
```
